```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("token_usage.csv")
```

```python
df["task_id"] = df["task_id"].replace("20250215-120903", "survery Large Language Model papers")
```

```python
df
```

```python
s = df.groupby(by=["task_id","model"]).sum()
```

```python
type(s)
```

```python
s= s.reset_index(drop=False)
```

```python
s
```

```python
def cost_function(row):
    if (row["model"] == "gpt-4o-mini"):
        return row["prompt_tokens"]* 0.15/1000000.0 + 0.6 /1000000.0 * row["completion_tokens"]
    else:
        return row["prompt_tokens"]* 2.5/1000000.0 + 10.0 /1000000.0 * row["completion_tokens"]
```

```python
s["cost"] = s.apply(cost_function, axis= 1)
```
